```python
import cv2 as cv
import matplotlib.pyplot as plt
import matplotlib.colors as clr
import numpy as np
```

```python
img = cv.imread('Red_cat.jpg')
```

### Cхема RGB

```python
plt.axis('off')
plt.imshow(img)
plt.show()
```

```python
img_RGB = cv.cvtColor(img, cv.COLOR_BGR2RGB)
plt.axis('off')
plt.imshow(img_RGB)
plt.show()
```

```python
R = img_RGB.copy()
R[:,:,(1,2)] = 0
G = img_RGB.copy()
G[:,:,(0,2)] = 0
B = img_RGB.copy()
B[:,:,(0,1)] = 0
```

```python
plt.axis('off')
plt.imshow(R)
plt.show()
```

```python
plt.axis('off')
plt.imshow(G)
plt.show()
```

```python
plt.axis('off')
plt.imshow(B)
plt.show()
```

```python
new_RGB = R + G + B
```

```python
plt.axis('off')
plt.imshow(new_RGB)
plt.show()
```
